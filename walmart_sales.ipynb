```python
# Convert the "Date" column to datetime format with correct format specification
df["Date"] = pd.to_datetime(df["Date"], format="%d-%m-%Y")
```

```python
# Then extract the month name using the .dt accessor
df["Month"] = df["Date"].dt.month_name()
```

```python
df.head()
```

```python
monthly_sales=df.groupby("Month")["Weekly_Sales"].sum()
monthly_sales
```

```python
highest_month = monthly_sales.idxmax()
highest_month 
```

```python
highest_sales = monthly_sales.max()
highest_sales
```

```python
#sales by store sales
sales_store=df.groupby("Store")["Weekly_Sales"].sum()
sales_store
```

```python
store_number=sales_store.idxmax()
store_number
```

```python
store_sales=sales_store.max()
store_sales
```
